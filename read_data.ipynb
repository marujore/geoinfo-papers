```python
from pathlib import Path

import pandas as pd
import numpy as np
```

```python
geoinfo_dir = Path('/home/gabriel-david/Estudo/Mestrado/cienciadados/geoinfo-papers/2013')
```

```python
artigos = pd.read_csv(
    geoinfo_dir / 'artigos.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Artigos: {len(artigos)} registros")

pessoas = pd.read_csv(
    geoinfo_dir / 'pessoas.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Pessoas: {len(pessoas)} registros")

autoria = pd.read_csv(
    geoinfo_dir / 'autoria.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Autoria: {len(autoria)} registros")

comite = pd.read_csv(
    geoinfo_dir / 'comite_cientifico.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Comitê: {len(comite)} registros")

organizacao = pd.read_csv(
    geoinfo_dir / 'organização.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Organização: {len(organizacao)} registros")

evento = pd.read_csv(
    geoinfo_dir / 'evento.csv',
    quotechar='"',
    quoting=1,
    encoding='utf-8-sig'
)
print(f"Evento: {len(evento)} registros")
```

```python
#DF Papers
df = artigos.merge(autoria, on='id_artigo')
df = df.merge(pessoas[['id_pessoa', 'nome', 'instituicao']], on='id_pessoa')

df_papers = df.groupby('id_artigo').agg({
    'titulo': 'first',
    'abstract': 'first',
    'tipo_artigo': 'first',
    'lingua': 'first',
    'n_edicao': 'first',
    'nome': list,
    'instituicao': list
}).reset_index()
df_papers
```

```python
#DF Comitê
df_comite = comite.merge(pessoas, on='id_pessoa')
df_comite = df_comite[['id_pessoa', 'nome', 'instituicao', 'n_edicao']]
df_comite = df_comite.sort_values('nome').reset_index(drop=True)
df_comite
```
